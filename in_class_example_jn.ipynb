#### Iris Plot 

```python
file_name2 = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
iris = pd.read_csv(file_name2)
iris["species"] = pd.Categorical(iris["species"])

plt.rcParams["figure.figsize"] = (8,8)
fig, ax = plt.subplots()
iris.plot.scatter(x="sepal_length", y="sepal_width", c="species",marker= '^', cmap="viridis", s=50, grid="TRUE", ax=ax);
```
